# Ciclo de Seminários Técnicos sobre avanços em Sistemas de Comunicações 

### Objetivos
As metas desse trabalho são:
- Colocar o aluno em contato com material técnico sobre Sistemas Modernos de Comunicação 
- Incentivar a preparação de material técnico pelo aluno
- Avaliar as habilidades de entendimento de leitura técnica, objetividade na apresentação de assuntos técnicos

### Objeto da avaliação

Sistemas de comunicações móveis são evoluídos de forma a gerar uma nova geração de sistemas a cada 10 anos. Por isso, o engenheiro de telecomunicações precisa ser um profissional em constante atualização técnico-científica. Esse projeto almeja confrontar o aluno a temas de vanguarda na área de telecomunicações, avaliando suas habilidades de entendimento de leitura técnica e objetividade na apresentação de assuntos técnicos na produção de conteúdo em vídeo. Assim, sua entrega será a produção de um vídeo técnico de caráter informativo sobre um dos temas proposto pelo docente da disciplina. 

### Entregas

Um vídeo com as seguintes características:
 - Duração: entre 5 e 10 minutos;
 - Publicação no youtube: mandar o link via SIGAA;
 - Todo slide do vídeo deverá conter as referências utilizadas (seja para figuras ou textos);
 - Figuras e animações próprias são bem-vindas;
 - O vídeo deve ser narrado pelo aluno. 

### Avaliação
- Dois alunos irão ser escolhidos para a avaliação dos vídeos dos colegas;
- O professor fará a avaliação dos vídeos de cada aluno;
- Um membro externo a disciplina também avaliará o vídeo. Quem sabe alguém do mercado!
- A pontuação será a média das notas atribuídas ao vídeo.

### Prazo de entrega do vídeo
- 10/02/2022 (envio do link do vídeo via SIGAA)
